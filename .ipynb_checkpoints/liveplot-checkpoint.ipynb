```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
import csv
from matplotlib.animation import FuncAnimation
```

```python
liveData = pd.read_csv("live.csv", usecols=[0,1,2,3,4],names=['time','ch1','ch2','ch3','ch4'], header=None)
liveData.shape
```

```python
fig = plt.figure()

#add ch
ax_ch1 = fig.add_subplot(2,2,1)
ax_ch1.set(xlim=[0.,2.], ylim=[-1200,1000], title='ch1')

ax_ch2 = fig.add_subplot(2,2,2)
ax_ch2.set(xlim=[0.,8.], ylim=[-1200,1000], title='ch2')

ax_ch3 = fig.add_subplot(2,2,3)
ax_ch3.set(xlim=[0.,8.], ylim=[-1200,1000], title='ch3')

ax_ch4 = fig.add_subplot(2,2,4)
ax_ch4.set(xlim=[0.,8.], ylim=[-1200,1000], title='ch4')
```

```python
def animate(i):
    #add ch
    liveData = pd.read_csv("live.csv", usecols=[0,1,2,3,4],names=['time','ch1','ch2','ch3','ch4'], header=None)
    x = liveData['time']
    
    #add ch
    y1 = liveData['ch1']
    y2 = liveData['ch2']
    y3 = liveData['ch3']
    y4 = liveData['ch4']
    
    #add ch
    ax_ch1.clear()
    ax_ch2.clear()
    ax_ch3.clear()
    ax_ch4.clear()
    
    
    #add ch
    ax_ch1.set(xlim=[0.+ i*0.5 ,2.+ i*0.5], ylim=[-1200,1000]) 
    
    #add ch
    ax_ch1.plot(x,y1)
    ax_ch2.plot(x,y2)
    ax_ch3.plot(x,y3)
    ax_ch4.plot(x,y4)
    
    

```

```python
ani = FuncAnimation(fig, func=animate, frames= np.arange(0,10,1), interval=3000) 
plt.show()
```
