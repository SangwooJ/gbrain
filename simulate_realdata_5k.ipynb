```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
import csv
```

```python
#load data
cols=[0,27,28,29]
labels=['time','ch27','ch28','ch29']
bdata = pd.read_csv("realdata.csv",usecols=cols, names=labels, header=None)
bdata.shape
```

```python
#bdata = bdata.drop_duplicates(['time'])
timeList=bdata['time'].values.tolist()

#add ch
ch1List=bdata['ch28'].values.tolist()
```

```python
def updataCSV():
    f = open('live.csv','a', newline='')
    wr = csv.writer(f)
    count = 500 # 500 = 0.1second
    for i in range(170):
        time_split = timeList[i*count:(i+1)*count] #i=1 -> 1000 = 0.1second
        
        #add ch
        ch1_split = ch1List[i*count:(i+1)*count]
        
        time.sleep(0.5)
        print("5000...done"+ '%.10f' % time_split[-1])
        
        #add ch
        for t,c1 in zip(time_split, ch1_split):
            #add ch
            wr.writerow([t,c1])
    f.close()
```

```python
updataCSV()
```
