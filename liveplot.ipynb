```python
fig = plt.figure(figsize=(5,5))

#add ch
ax_ch1 = fig.add_subplot(2,2,1)
ax_ch1.set(xlim=[0.,4.], ylim=[-500,500], title='ch1')

ax_ch2 = fig.add_subplot(2,2,2)
ax_ch2.set(xlim=[0.,4.], ylim=[-500,500], title='ch2')

ax_ch3 = fig.add_subplot(2,2,3)
ax_ch3.set(xlim=[0.,4.], ylim=[-500,500], title='ch3')

ax_ch4 = fig.add_subplot(2,2,4)
ax_ch4.set(xlim=[0.,4.], ylim=[-500,500], title='ch4')
```

```python
def animate(i):
    #add ch
    liveData = pd.read_csv("realdata.csv", usecols=[0,26,27,28,29],names=['time','ch1','ch2','ch3','ch4'], header=None)
    x = liveData['time']
    
    #add ch
    y1 = liveData['ch1']
    y2 = liveData['ch2']
    y3 = liveData['ch3']
    y4 = liveData['ch4']
    
    #add ch
    ax_ch1.clear()
    ax_ch2.clear()
    ax_ch3.clear()
    ax_ch4.clear()
    
    #add ch
    ax_ch1.set(xlim=[-1.+ i*0.25 ,0.+ i*0.25], ylim=[-1200,1500])
    ax_ch2.set(xlim=[-1.+ i*0.25 ,0.+ i*0.25], ylim=[-1200,1500]) 
    ax_ch3.set(xlim=[-1.+ i*0.25 ,0.+ i*0.25], ylim=[-1200,1500]) 
    ax_ch4.set(xlim=[-1.+ i*0.25 ,0.+ i*0.25], ylim=[-1200,1500]) 
    
    #add ch
    ax_ch1.plot(x,y1)
    ax_ch2.plot(x,y2)
    ax_ch3.plot(x,y3)
    ax_ch4.plot(x,y4)
    
    

```

```python
ani = FuncAnimation(fig, func=animate, frames= np.arange(0,100,1), interval=1000) 
plt.show()
```
